Feature Engineering for shipping days hypothesis

```python
from scipy.stats import chi2_contingency
def chi_square_test(table):
    stat, p, dof, expected = chi2_contingency(table)
    significance_level = 0.05
    if p <= significance_level:
        print('Dependent (reject H0)')
    else:
        print('Independent (accept H0)')
```

```python
# test if relations between each columns and traget column(Days for shipping (real)) is strong
column_list = days_pred_df.columns.tolist()
for column in column_list:
    if(column != "Days for shipping (real)"):
        # Create a contingency table
        contingency_table = pd.crosstab(days_pred_df[column], days_pred_df['Days for shipping (real)'])
        print(column)
        chi_square_test(contingency_table)
```

"Order Item Quantity" column makes sense from a domain perspective.it logically have an impact on shipping days. it is worth retaining the column even if it doesn't show a significant association in the chi-square test. Plus it will appear in user input so we do not drop any columns from the result of chi square test

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Convert data to PyTorch tensors
X_tensor = torch.tensor(X.values, dtype=torch.float32)
y_tensor = torch.tensor(y.values, dtype=torch.float32).unsqueeze(1)  # Add a dimension for regression

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)
```

```python
from SimpleNeuralNet import NeuralNetwork
```

```python
# Set hyperparameters
input_dim = X.shape[1]  # Total number of input features
hidden_dim = 64
# Instantiate the model
model = NeuralNetwork(input_dim, hidden_dim)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay = 0.001)

# Training loop
patience = 50
counter = 0
lowest_mse = 9999
num_epochs = 2500
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    y_pred = model(X_test)
    mse = mean_squared_error(y_test.detach().numpy(), y_pred.detach().numpy())
    if mse < lowest_mse:
        lowest_mse = mse
        counter = 0
    else:
        counter += 1
        if counter >= patience:
            print(f"Breaking early at epoch {epoch}, MSE:{mse}")
            break
    
    if (epoch + 1) % 500 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
# Evaluation
model.eval()
with torch.no_grad():
    y_pred = model(X_test)
    mse = mean_squared_error(y_test.numpy(), y_pred.numpy())
    print(f"Mean Squared Error: {mse:.4f}")
```
